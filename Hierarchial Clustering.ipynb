```python
import pandas as pd
data = pd.read_csv('Mall_Customers.csv')
```

```python
shoping = data[:100].iloc[:,3:5].values
```

```python
xplot = sch.dendrogram(sch.linkage(shoping,method="single"))
plt.title('dendrogram')
plt.xlabel('x values')
plt.ylabel('Distance')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='single')
cluster.fit_predict(shoping)
```

```python
plt.scatter(shoping[:,0],shoping[:,1],c = cluster.labels_,cmap='rainbow')
```

```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
Mall_Customers = pd.read_csv('Mall_Customers.csv')
```

```python
data = Mall_Customers.iloc[:,3:5].values
```

```python
plt.scatter(data[:,0],data[:,1])
```

```python
Hierarchial = sch.dendrogram(sch.linkage(data,"single"))
plt.title('dendrogram')
plt.xlabel('x values')
plt.ylabel('Distance')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=10,affinity='euclidean',linkage='single')
cluster.fit_predict(data)
```

```python
plt.scatter(data[:,0],data[:,1],c = cluster.labels_,cmap='rainbow')
```
